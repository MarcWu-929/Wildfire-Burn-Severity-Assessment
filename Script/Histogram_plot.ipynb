```python
from osgeo import gdal
import matplotlib.pyplot as plt
import numpy as np

def plot_histogram(tiff_path):
    # Open the TIFF file
    dataset = gdal.Open(tiff_path)

    if dataset is None:
        print(f"Error: Unable to open {tiff_path}")
        return

    # Get the number of bands in the image
    num_bands = dataset.RasterCount

    # Plot histogram for each band
    for i in range(7, num_bands + 1):
        band = dataset.GetRasterBand(i)
        band_array = band.ReadAsArray()

        # Flatten the 2D array to 1D
        flat_array = band_array.flatten()

        # Plot histogram for 16-bit image
        plt.hist(flat_array, bins=np.arange(0, 4000), density=False, cumulative=False, color='tab:blue', alpha=0.75)
        # plt.title(f'Band {i} Histogram')
        plt.xlabel('Pixel Value')
        plt.ylabel('Frequency')
        plt.show()

if __name__ == "__main__":
    tiff_path = r"C:\Document\University\2023_Spring_Junior\GEOG_491\Dataset\August_Complex\Raw_training\Post\clipped_post.tif"
    plot_histogram(tiff_path)
```
